```python
# 选择拥有浴室的房源
baths = su[su['propertyinfo_value'].str.contains('ba')]
```

```python
# 使用项目符号进行切分
def parse_info(row):
    if not 'sqft' in row:
        bd, ba = row.split('•')[:2]
        sqft = np.nan
    else:
        bd, ba, sqft = row.split('•')[:3]
    return pd.Series({'Beds':bd, 'Baths':ba, 'Sqft':sqft})
```

```python
attr = baths['propertyinfo_value'].apply(parse_info)
```

```python
attr.head()
```

```python
#去除字符串标签,strip()用于移除字符串首尾空格
attr_cln = attr.applymap(lambda x: x.strip().split(' ')[0] if isinstance(x, str) else np.nan)
attr_cln.head()
```

```python
# 与原表连接
sujnd = baths.join(attr_cln)
sujnd.T
```

```python
# 分析出邮编与楼层
def parse_addy(r):
    so_zip = re.search(', NY(\d+)', r)
    so_flr = re.search('(?:APT|#)\s+(\d+)[A-Z]+,', r)
    if so_zip:
        zipc = so_zip.group(1)
    else:
        zipc = np.nan
    if so_flr:
        flr = so_flr.group(1)
    else:
        flr = np.nan 
    return pd.Series({'Zip':zipc, 'Floor':flr})
```

```python
flrzip = sujnd['routable_link/_text'].apply(parse_addy)
flrzip.head()
```

```python
len(flrzip[~flrzip['Floor'].isnull()])
```

```python
len(flrzip[~flrzip['Zip'].isnull()])
```

```python
suf = sujnd.join(flrzip)
suf.T
```

```python
# 选出需要的列
sudf = suf[['pricelarge_value_prices', 'Beds', 'Baths', 'Sqft', 'Floor', 'Zip']]
```

```python
# 重命名
# inplace=True：不创建新的对象，直接对原始对象进行修改
# inplace=False：对数据进行修改，创建并返回新的对象承载其修改结果
sudf.rename(columns={'pricelarge_value_prices':'Rent'}, inplace=True)
sudf.head()
```

```python
# 重置索引
# drop=True：在原有的索引列重置索引，不再另外添加新列。
# drop=False：原有的索引不变添加列名index，同时在新列上重置索引
sudf.reset_index(drop=True, inplace=True)
sudf.head()
```

```python
sudf.describe()
```

```python
# 将Studio变为0
sudf.loc[:,'Beds'] = sudf['Beds'].map(lambda x: 0 if 'Studio' in x else x)
sudf.head()
```

```python
sudf.info()
```

```python
sudf.loc[:, 'Rent'] = sudf["Rent"].astype(int)
sudf.loc[:, 'Beds'] = sudf['Beds'].astype(int)
sudf.loc[:, 'Baths'] = sudf['Baths'].astype(float)

sudf.loc[:, 'Sqft'] = sudf['Sqft'].str.replace(',', '')
sudf.loc[:, 'Sqft'] = sudf['Sqft'].astype(float)
sudf.loc[:, 'Floor'] = sudf['Floor'].astype(float)
```

## Regression Modeling

```python
sudf2.head()
```

```python
sudf2.sort_values('Zip').head()
```

```python
import patsy
import statsmodels.api as sm
```

```python
f = 'Rent ~ Zip + Beds'
y, X = patsy.dmatrices(f, sudf2, return_type='dataframe')
```

```python
results = sm.OLS(y, X).fit()
print(results.summary())
```

```python
X.head()
```

```python
# pd.DataFrame(data=None, index=None, columns=None,
#              dtype=None, copy=False)
to_pred_idx = X.iloc[0].index    # 提取第一行索引
to_pred_zero = np.zeros(len(to_pred_idx))   # 生成0矩阵
tpdf = pd.DataFrame(data=to_pred_zero, index=to_pred_idx, columns=['value'])
```

```python
tpdf
```

```python
tpdf.loc['Intercept'] = 1
tpdf.loc['Beds'] = 1
tpdf.loc['Zip[T.10009]'] = 1
tpdf
```
